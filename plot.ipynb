```python
import pandas as pd

df = pd.read_excel('gpu_pregel_eval.xlsx')
```

```python
# remove row 5, 10, 14
df = df.drop([4, 9, 14])
df
```

```python
# new df of the first 4 rows without index column
df1 = df.iloc[:4, :4]
df1 = df1.transpose()
# export to latex with no index
print(df1.to_latex(index=False))
```

```python
# new dataframe of the remaining rows without index column
df2 = df.iloc[4:, :4]
# export to latex with no index
print(df2.to_latex(index=False))
```

```python
df2
```

```python

import matplotlib.pyplot as plt
import numpy as np

titles = df2.iloc[:,0].to_list()
values_notredame = df2.iloc[:,1].to_list()
values_berkstan = df2.iloc[:,2].to_list()
values_livejournal = df2.iloc[:,3].to_list()

# Set up the bar width and positions
bar_width = 0.25
r1 = np.arange(len(titles))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# plot as bar chart
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(r1, values_notredame, label='web-NotreDame', width=bar_width)
ax.bar(r2, values_berkstan, label='web-BerkStan', width=bar_width)
ax.bar(r3, values_livejournal, label='soc-LiveJournal', width=bar_width)
ax.set_ylabel('Time in ms')
ax.legend()
plt.xticks([r + bar_width for r in range(len(titles))], titles, rotation=45, ha='right')
```
